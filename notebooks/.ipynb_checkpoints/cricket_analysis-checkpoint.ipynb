```python
batsman_stats_df = batsman_df[['Player Name','Batting Average','Highest Score','Centuries',
                               'Centuries','Fifties','Ducks','Fours','Sixes']].sort_values(by='Batting Average',
                                                                                           ascending=False)
    
batsman_stats_df.head()
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(23,22))
sns.barplot(data=batsman_stats_df,y='Player Name',x='Batting Average')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort values for better visualization
# batsman_stats_df = batsman_stats_df.sort_values(by='Batting Average', ascending=False)

# Style
sns.set_style("whitegrid")

plt.figure(figsize=(12, 10))

sns.barplot(
    data=batsman_stats_df,
    y='Player Name',
    x='Batting Average',
    palette='viridis'
)

plt.title("Batting Average by Player", fontsize=16)
plt.xlabel("Batting Average", fontsize=12)
plt.ylabel("Player Name", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(12, 10))

sns.barplot(
    data=batsman_stats_df,
    x='Player Name',
    y='Highest Score',
    palette='viridis',
)
```

```python
top10 = batsman_stats_df.sort_values(by='Batting Average', ascending=False).head(10)
```

```python
top10
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(8, 6))

sns.barplot(
    data=top10.sort_values(by='Highest Score',ascending=False),
    y='Player Name',
    x='Highest Score',
    palette='tab20b',
)


plt.title("Top-10 Highest Score by Players", fontsize=16)
plt.xlabel("Highest Score", fontsize=12)
plt.ylabel("Player Name", fontsize=12)

plt.tight_layout()
plt.show()
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(8, 6))

sns.barplot(
    data=batsman_stats_df.sort_values(by='Sixes', ascending=False).head(20),
    x='Player Name',
    y='Fours',
    palette='tab20b',
    
)


plt.title("Top-10 Highest Score by Players", fontsize=16)
plt.xlabel("Player Name", fontsize=12)
plt.ylabel("Fours", fontsize=12)

# Rotate x-axis labels
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
sns.set_style("whitegrid")

# Select top 20 players by Sixes
df = batsman_stats_df.sort_values(by='Sixes', ascending=False).head(20)

# Convert to long format
df_melted = df.melt(
    id_vars='Player Name',
    value_vars=['Fours', 'Sixes'],
    var_name='Type',
    value_name='Count'
)

plt.figure(figsize=(12,6))

sns.barplot(
    data=df_melted,
    x='Player Name',
    y='Count',
    hue='Type',
    palette='CMRmap_r'
)

plt.title("Fours vs Sixes by Players", fontsize=16)
plt.xlabel("Player Name")
plt.ylabel("Count")

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

fig, axes = plt.subplots(2, 2, figsize=(16,10))

# 1️⃣ Champion Count
sns.countplot(
    data=champion_df,
    x="Champion",
    order=champion_df["Champion"].value_counts().index,
    palette="viridis",
    ax=axes[0,0]
)
axes[0,0].set_title("Asia Cup Titles by Team")


# 2️⃣ Host Distribution
sns.countplot(
    data=champion_df,
    x="Host",
    order=champion_df["Host"].value_counts().index,
    palette="Set2",
    ax=axes[0,1]
)
axes[0,1].set_title("Asia Cup Host Countries")


# 3️⃣ Player of the Series
top_players = champion_df["Player Of The Series"].value_counts().head(5)

sns.barplot(
    x=top_players.values,
    y=top_players.index,
    palette="magma",
    ax=axes[1,0]
)
axes[1,0].set_title("Top Player of the Series Winners")


# 4️⃣ Teams Growth
sns.lineplot(
    data=champion_df,
    x="Year",
    y="No Of Team",
    marker="o",
    ax=axes[1,1]
)
axes[1,1].set_title("Tournament Expansion Over Time")


plt.suptitle("Asia Cup Tournament Dashboard", fontsize=18)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

fig, axes = plt.subplots(3, 2, figsize=(20,15))

# 1️⃣ Champion Share (Pie Chart)
champion_df["Champion"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90,
    ax=axes[0,0],
    cmap="Set2_r"
)
axes[0,0].set_title("Asia Cup Titles Share")
axes[0,0].set_ylabel("")


# 2️⃣ Champion Count
sns.countplot(
    data=champion_df,
    x="Champion",
    order=champion_df["Champion"].value_counts().index,
    palette="viridis",
    ax=axes[0,1]
)
axes[0,1].set_title("Asia Cup Titles by Team")


# 3️⃣ Host Distribution
sns.countplot(
    data=champion_df,
    x="Host",
    order=champion_df["Host"].value_counts().index,
    palette="Set2",
    ax=axes[1,0]
)
axes[1,0].set_title("Asia Cup Host Countries")


# 4️⃣ Player of the Series
top_players = champion_df["Player Of The Series"].value_counts().head(5)

sns.barplot(
    x=top_players.values,
    y=top_players.index,
    palette="magma",
    ax=axes[1,1]
)
axes[1,1].set_title("Top Player of the Series Winners")


# 5️⃣ Teams Growth
sns.lineplot(
    data=champion_df,
    x="Year",
    y="No Of Team",
    marker="o",
    ax=axes[2,0]
)
axes[2,1].set_title("Tournament Expansion Over Time")


# 6️⃣ Empty space (for balance)
axes[2,1].axis("off")

plt.suptitle("Asia Cup Tournament Dashboard", fontsize=20)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style("whitegrid")

fig, axes = plt.subplots(2, 2, figsize=(16,10))

# 1️⃣ Champion Count
sns.countplot(
    data=champion_df,
    x="Champion",
    order=champion_df["Champion"].value_counts().index,
    palette="viridis",
    ax=axes[0,0]
)
axes[0,0].set_title("Asia Cup Titles by Team")


# 2️⃣ Host Distribution
sns.countplot(
    data=champion_df,
    x="Host",
    order=champion_df["Host"].value_counts().index,
    palette="Set2",
    ax=axes[0,1]
)
axes[0,1].set_title("Asia Cup Host Countries")


# 3️⃣ Player of the Series
top_players = champion_df["Player Of The Series"].value_counts().head(5)

sns.barplot(
    x=top_players.values,
    y=top_players.index,
    palette="magma",
    ax=axes[1,0]
)
axes[1,0].set_title("Top Player of the Series Winners")


# 4️⃣ Teams Growth
sns.lineplot(
    data=champion_df,
    x="Year",
    y="No Of Team",
    marker="o",
    ax=axes[1,1]
)
axes[1,1].set_title("Tournament Expansion Over Time")


plt.suptitle("Overall Asia Cup Tournament Analysis Report", fontsize=18)

plt.tight_layout()
plt.show()
sns.set_style("darkgrid")

fig, axes = plt.subplots(3, 2, figsize=(18,18))

# 1️⃣ Champion Share
champion_df["Champion"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90,
    ax=axes[0,0]
)
axes[0,0].set_title("Asia Cup Titles Share")
axes[0,0].set_ylabel("")


# 2️⃣ Top Wicket Takers
sns.barplot(
    data=bowler_df.sort_values(by='Wickets', ascending=False).head(10),
    x='Player Name',
    y='Wickets',
    palette='tab20b',
    ax=axes[0,1]
)
axes[0,1].set_title("Top Wicket Takers")
axes[0,1].tick_params(axis='x', rotation=90)


# 3️⃣ Best Economy Bowlers
sns.barplot(
    data=bowler_df.sort_values(by='Economy Rate', ascending=True).head(10),
    x='Player Name',
    y='Economy Rate',
    palette='coolwarm',
    ax=axes[1,0]
)
axes[1,0].set_title("Best Bowling Economy")
axes[1,0].tick_params(axis='x', rotation=90)


# 4️⃣ Five Wicket Hauls
sns.barplot(
    data=bowler_df.sort_values(by='Five Wickets', ascending=False).head(7),
    x='Player Name',
    y='Five Wickets',
    palette='pink',
    ax=axes[1,1]
)
axes[1,1].set_title("Most 5-Wicket Hauls")
axes[1,1].tick_params(axis='x', rotation=90)


# 5️⃣ Fours vs Sixes
df = batsman_stats_df.sort_values(by='Sixes', ascending=False).head(10)

df_melted = df.melt(
    id_vars='Player Name',
    value_vars=['Fours', 'Sixes'],
    var_name='Type',
    value_name='Count'
)

sns.barplot(
    data=df_melted,
    x='Player Name',
    y='Count',
    hue='Type',
    palette='CMRmap_r',
    ax=axes[2,0]
)

axes[2,0].set_title("Fours vs Sixes")
axes[2,0].tick_params(axis='x', rotation=90)


# 6️⃣ Highest Individual Scores
sns.barplot(
    data=top10.sort_values(by='Highest Score', ascending=False),
    y='Player Name',
    x='Highest Score',
    palette='tab20b',
    ax=axes[2,1]
)

axes[2,1].set_title("Highest Individual Scores")


# plt.suptitle("Asia Cup Cricket Analytics Dashboard", fontsize=20)

plt.tight_layout()
```
